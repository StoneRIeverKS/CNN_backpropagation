# 複数の入力を持つ勾配降下法による学習

```julia
using LinearAlgebra

# 特徴量
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

# 目的変数
win_or_lose_binary = [1, 1, 0, 1]

# 重みの初期値
weights = [0.1, 0.2, -0.1]

# alpha
alpha = 0.01

# 特徴量を1つのベクトルにまとめる
input = [toes[1], wlrec[1], nfans[1]]

# 1回だけ重みを更新
## 予測値を計算する
pred = dot(weights, input)

## deltaを計算する
delta = pred-win_or_lose_binary[1]

## 更新前のError
err = delta^2
println("更新前のError:", err)
## inputをかける
weights_delta = delta*input

## 重みを更新する
weights = weights-alpha*weights_delta

## 更新後のError
pred = dot(weights, input)
println("更新後のError", (pred-win_or_lose_binary[1])^2)

# ちゃんとErrorが減っている
```

```julia
using LinearAlgebra

# 特徴量
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

# 目的変数
win_or_lose_binary = [1, 1, 0, 1]

# 重みの初期値
weights = [0.1, 0.2, -0.1]

# alpha
alpha = 0.01

# 特徴量を1つのベクトルにまとめる
input = [toes[1], wlrec[1], nfans[1]]

for i = 1:3
    # 1回だけ重みを更新
    ## 予測値を計算する
    pred = dot(weights, input)

    ## deltaを計算する
    delta = pred-win_or_lose_binary[1]

    ## inputをかける
    weights_delta = delta*input

    ## 重みを更新する
    weights = weights-alpha*weights_delta

    ## 更新後のError
    pred = dot(weights, input)
    println("更新後のError", (pred-win_or_lose_binary[1])^2)
end

# ちゃんとErrorが減少している
```

# 一つの重みを凍結する

```julia
using LinearAlgebra

# 特徴量
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

# 目的変数
win_or_lose_binary = [1, 1, 0, 1]

# 重みの初期値
weights = [0.1, 0.2, -0.1]

# alpha
alpha = 0.3

# 特徴量を1つのベクトルにまとめる
input = [toes[1], wlrec[1], nfans[1]]

for i = 1:3
    # 1回だけ重みを更新
    ## 予測値を計算する
    pred = dot(weights, input)

    ## deltaを計算する
    delta = pred-win_or_lose_binary[1]

    ## inputをかける
    weights_delta = delta*input
    weights_delta[1] = 0 # 一つの重みを凍結する

    ## 重みを更新する
    weights = weights-alpha*weights_delta

    ## 更新後のError
    pred = dot(weights, input)
    println("更新後のError", (pred-win_or_lose_binary[1])^2)
end

# ちゃんとErrorが減少している!1つのパラメータは更新していないのに、Errorがちゃんと減少する
```

　上記についての補足  
　損失関数は、下記で表される。
$$L = (\omega_1 x_1+\omega_2 x_2+\omega_3 x_3-y)^2$$
　したがって、$L$は$\omega_2, \omega_3$に依存している。したがって、$\omega_2, \omega_3$が適切に更新されれば$\omega_1$が更新されなくても、損失関数$L$は減少することになる。  
　ただし、ある程度は損失関数の減少はあるだろうが、近似的な減少になると考えられる。

# 複数の出力を持つ勾配降下法による学習

```julia
# 特徴量
wlrec = [0.65, 1.0, 1.0, 0.9]

# 目的変数
hurt = [0.1, 0.0, 0.0, 0.1]
win = [1, 1, 0, 1]
sad = [0.1, 0.0, 0.1, 0.2]

# ニューラルネットワークの入力
input = wlrec[1]

# ニューラルネットワークの出力
truth = [hurt[1], win[1], sad[1]]

# パラメータの初期値
weights = [0.3, 0.2, 0.9]

# 学習率
alpha = 0.01

# 1回だけ更新
## 予測値の計算
pred = input*weights

## deltaの計算
delta = pred-truth

## weights_deltaの計算
weights_delta = input*delta

## weightsの更新
weights -= weights_delta*alpha

## Errorの計算と表示
delta = input*weights-truth
for d in delta
    println("Error:", d^2)
end
println("---")
```

# 複数の入力と出力を持つ勾配降下法

```julia
using LinearAlgebra

# 特徴量
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

# 目的変数
hurt = [0.1, 0.0, 0.0, 0.1]
win = [1, 1, 0, 1]
sad = [0.1, 0.0, 0.1, 0.2]

# ニューラルネットワークの入力
input = [toes[1], wlrec[1], nfans[1]]

# ニューラルネットワークの出力
truth = [hurt[1], win[1], sad[1]]

# 重みの初期値
weights = [0.1 0.1 -0.3; 0.1 0.2 0.0; 0.0 1.3 0.1]

# 学習率
alpha = 1

# 1回だけ更新
## pred_vecの計算
pred_vec = weights*input

## delta_vecの計算
delta_vec = pred_vec-truth

## weights_delta_matrixの計算
weights_delta_matrix = zeros(size(input)[1], size(delta_vec)[1])

for i = 1:size(input)[1]
    for j = 1:size(delta_vec)[1]
        setindex!(weights_delta_matrix, input[i]*delta_vec[j], i, j)
    end
end

## 更新
weights -= alpha*weights_delta_matrix

## Errorの計算
pred_vec = weights*input
for delta in (pred_vec-truth)
    println(delta^2)
end

println("---")
```

# これらの重みは何を学習するか

```julia
import Pkg
# 必要なパッケージのinstall
Pkg.add("MLDatasets")
Pkg.add("Flux")
```

```julia
using MLDatasets
x_train, y_train = MLDatasets.MNIST.traindata(Float32)
```

```julia
# 平坦化
using Flux
using Flux.Data: DataLoader
using Flux: onehotbatch, onecold

x_train = Flux.flatten(x_train)
y_train = onehotbatch(y_train, 0:9)
```

```julia
# ニューラルネットワークの入力
input = x_train[:, 1]

# ニューラルネットワークの出力
truth = y_train[:, 1]

# 重みの初期値
weights = zeros(size(y_train)[1], size(x_train)[1], )

# 学習率
alpha = 0.005

# 10回だけ更新
for i = 1:100
    ## pred_vecの計算
    pred_vec = weights*input

    ## delta_vecの計算
    delta_vec = pred_vec-truth

    ## weights_delta_matrixの計算
    weights_delta_matrix = zeros(size(delta_vec)[1], size(input)[1])

    for i = 1:size(delta_vec)[1]
        for j = 1:size(input)[1]
            setindex!(weights_delta_matrix, delta_vec[i]*input[j], i, j)
        end
    end

    ## 更新
    weights -= alpha*weights_delta_matrix

    ## Errorの計算
    pred_vec = weights*input
    
    for delta in (pred_vec-truth)
        println(delta^2)
    end

    println("---")
    
end
# ちゃんとErrorは減少した
```
